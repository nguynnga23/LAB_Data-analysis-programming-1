### EX3:	Get the proportion of employees by Education level from attrition_pop.
### •	Use *proportional stratified sampling* on attrition_pop to sample 40% of each Education group, setting the seed to 2022.
### •	Get the proportion of employees by Education level from attrition_strat.

```python
# Get the proportion of employees by Education level from attrition_pop
proportion_education_pop = attrition_pop["Education"].value_counts(normalize=True)
print(proportion_education_pop)
```

```python
# Use proportional stratified sampling on attrition_pop to sample 40% of each Education group, setting the seed to 2022.
attrition_strat = attrition_pop.groupby("Education").sample(frac=0.4, random_state=2022)
attrition_strat
```

```python
# Get the proportion of employees by Education level from attrition_strat.
proportion_education_strat = attrition_strat["Education"].value_counts(normalize=True)
print(proportion_education_strat)
```

### EX4: Create a list of unique JobRole values from attrition_pop, and assign to job_roles_pop Randomly sample four JobRole values from job_roles_pop. Subset attrition_pop for the sampled job roles by filtering for rows where JobRole is in job_roles_samp.Remove any unused categories from JobRole.For each job role in the filtered dataset, take a random sample of ten rows, setting the seed to 2022.

```python
# Create a list of unique JobRole values from attrition_pop, and assign to job_roles_pop
job_roles_pop = list(attrition_pop["JobRole"].unique())
job_roles_pop
```

```python
# Subset attrition_pop for the sampled job roles by filtering for rows where  is in job_roles_samp
JobRole_condition = attrition_pop["JobRole"].isin(job_roles_samp)
attrition_cluster = attrition_pop[JobRole_condition]
attrition_cluster["JobRole"]
```

```python
# Remove any unused categories from JobRole.
# Convert the Column to Categorical:
attrition_cluster["JobRole"] = attrition_cluster["JobRole"].astype("category")
attrition_cluster["JobRole"] = attrition_cluster["JobRole"].cat.remove_unused_categories()
```

```python
# For each job role in the filtered dataset, take a random sample of ten rows, setting the seed to 2022.
attrition_cluster = attrition_cluster.groupby("JobRole").sample(n=10, random_state=2022)
attrition_cluster["JobRole"].head(10)
```

### EX5: Load the dataset and find the most common gender and birth country.
### Load the dataset into a DataFrame using pandas and then extract the top values from sex and birth_country.

```python
# Load the dataset and find the most common gender and birth country
laureates_pop = pd.read_csv("nobel_laureates_data.csv")
gender_mode = laureates_pop["gender"].mode()[0]
print("The mode common gender:", gender_mode)
birth_country_mode = laureates_pop["bornCountry"].mode()[0]
print("The mode common birth country:", birth_country_mode)

```

```python
# Load the dataset into a DataFrame using pandas and then extract the top values from sex and birth_country.
laureates_pop['gender'] = laureates_pop['gender'].astype('category')
laureates_pop['bornCity'] = laureates_pop['bornCity'].astype('category')
gender_top = laureates_pop['gender'].value_counts().nlargest(n = 5)
birth_country_top = laureates_pop['bornCountry'].value_counts().nlargest(n = 6)
print("The top values from sex:\n", gender_top)
print("\nThe top values form birth country:", birth_country_top)
```

### EX6: Identify the decade with the highest proportion of US-born winners
### To calculate the proportion, first create a column that creates a flag for winners whose birth country is "United States of America", then create a decade column, and use both to find the proportion.

```python
# Identify the decade with the highest proportion of US-born winner

# 1. Filter US winners
us_winners = laureates_pop[laureates_pop["bornCountry"] == "USA"]
# 2. Add a "Decade" column
us_winners['Decade'] = (us_winners['year'] // 10) * 10
us_winners
# 3. Caculate proportion of US winners per decade
decade_proportions = (us_winners.groupby('Decade')['year'].count()/len(us_winners)) * 100
# print(us_winners.groupby('Decade')['year'].count())
# 4. The highest proportion of US-born winner
highest_decade = decade_proportions.idxmax()
highest_proportion = decade_proportions.max()
# print(decade_proportions)
print("The decade with the highest proportion of US-born winner: ", highest_decade)
print(f"Proportion of US winners in that decade: {highest_proportion:.2f}%\n")
```

### EX7: Find the decade and category with the highest proportion of female laureates
### You can copy and modify your code from the previous tasks to create a DataFrame for the proportion of female winners, then create a dictionary called max_female_dict with the year and category pair with the most female winners.

```python
# 1. Filter female winners 
female_winner = laureates_pop[laureates_pop["gender"] == 'female']

# 2. Create 'Decade' column
female_winner['decade'] = (laureates_pop['year']//10)*10 #Năm bắt đầu của thập kỉ

# 3.Calculate proportion per decade and category
decade_proportions = (female_winner.groupby(['decade', 'category'])['year'].count()/len(female_winner))*100

# 4. Find the maximum proportion (unstacking for clarity)
highest_decade = decade_proportions.idxmax()
highest_proportion = decade_proportions.max()

# 5. Create a dictionary with results
max_female_dict = {'Year': highest_decade[0], 'Category': highest_decade[1], 'Proportion': highest_proportion}

print("Dictionary with results:")
print(max_female_dict, "\n")
```
