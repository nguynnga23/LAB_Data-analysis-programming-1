### 3. Bài tập: Cho các số a, b, m, n trong đó a < b là hai số thực bất kỳ; m, n là các số nguyên dương. Viết hàm số tạo một mảng hai chiều có shape = (m, n), các phần tử là các số ngẫu nhiên phân bố đều trong nửa đoạn (a, b).



```python
# hàm số tạo một mảng hai chiều có shape = (m, n), các phần tử là các số ngẫu nhiên phân bố đều trong nửa đoạn (a, b).
def create_random_2d_array(a, b, m, n):
    return np.random.uniform(low=a, high=b, size=(m, n))

random_2d_array = create_random_2d_array(-3, 3, 4, 5)
random_2d_array
```

### 5. Generating random numbers
### - Generate 5000 numbers from a uniform distribution, setting the parameters low to -3 and high to 3.
### - Generate 5000 numbers from a normal distribution, setting the parameters loc to 5 and scale to 2.
### - Plot a histogram of uniforms with bins of width of 0.25 from -3 to 3 using plt.hist().
### - Plot a histogram of normals with bins of width of 0.5 from -2 to 13 using plt.hist().

```python
# Generate 5000 numbers from a uniform distribution, setting the parameters low to -3 and high to 3.
uniform_numbers = np.random.uniform(low=-3, high=3, size=5000)
uniform_numbers
```

```python
# Generate 5000 numbers from a normal distribution, setting the parameters loc to 5 and scale to 2.
normal_numbers = np.random.normal(loc=5, scale=2, size=5000)
normal_numbers
```

```python
# Plot a histogram of uniforms with bins of width of 0.25 from - 3 to 3 using plt.hist()
plt.hist(uniform_numbers, bins=np.arange(-3, 3.25, 0.25), edgecolor='black', color="pink")
plt.title("Uniform Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot a histogram of normals with bins of width of 0.5 from -2 to 13 using plt.hist().
plt.hist(normal_numbers, bins=np.arange(-2, 13.5, 0.5), edgecolor='black', color="green")
plt.title("Normal Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### 6. Nghiên cứu các phương pháp lấy mẫu, thực hành trên dataset data_coffee

### - simple : lấy mẫu ngâu nhiên đơn giản
### - systematic : Lấy mẫu ngẫu nhiên có hệ thống
### - stratified : Lấy mẫu ngẫu nhiên phần tầng
### - cluster : Lấy mẫu ngẫu nhiên theo cụm

```python
coffee_data = pd.read_csv("./data_coffee.csv")
coffee_data.head(5)
```

```python
# Simple sampling
coffee_data.sample(5)
```

```python
# Systematic sampling function
def systematic_sampling_coffee(coffee_data, step):
    indexes = np.arange(0, len(coffee_data), step = step)
    systematic_sample_coffee = coffee_data.iloc[indexes]
    return systematic_sample_coffee

systematic_sample_coffee = systematic_sampling_coffee(coffee_data, 50)
display(systematic_sample_coffee)
```

```python
# Stratified Sampling
coffee_data.groupby("Color", group_keys=False).apply(lambda x: x.sample())
```

```python
# Cluster Sampling
# step 01 - definition of variables
length = len(coffee_data)
groups = 10
elements = length / groups
print('elements by group:',elements)

# step 02 - group list creation
groups_list = []
groups_id = 0
elements_count = 0

for _ in coffee_data.iterrows():
    groups_list.append(groups_id)
    elements_count += 1
    if elements_count > elements:
        elements_count = 0
        groups_id += 1

np.unique(groups_list, return_counts=True)
np.shape(groups_list), coffee_data.shape

# step 03 - allocating the elements in their respective groups randomly
coffee_data['group'] = groups_list
display(coffee_data.head(3))

# step 04 - randomly choosing the group
import random as rd
rd.seed(None)
group_selected = rd.randint(0, groups - 1)

```
