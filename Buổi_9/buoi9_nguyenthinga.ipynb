### Q1: MÔ TẢ DỮ LIỆU

```python
import pandas as pd

schools = pd.read_csv("schools.csv")

# Mô tả dữ liệu
print(schools.head())
print(schools.shape)
print(schools.describe())
print(schools.isna().sum())
```

### Q2: Những trường có điểm Toán tốt nhất là trường mà điểm toán trung bình đạt 80% điểm tối đa (800 điểm). Viết đoạn code xuất ra danh sách các trường có điểm toán tốt nhất, lưu trong DataFrame tên best_math_schools ( gồm các cột "school_name" và  "average_math" , sắp xếp  "average_math"  theo thứ tự giảm dần.)

```python
# Tìm danh sách các trường có điểm toán trung bình đạt 80% vào DataFrame với 2 cột school_name và average_math
best_math_schools = schools[schools['average_math'] > 800*0.8]
best_math_schools = best_math_schools[['school_name', 'average_math']]

# Sắp xếp cột average_math theo thứ tự giảm dần
best_math_schools = best_math_schools.sort_values(by = 'average_math', ascending=False)
best_math_schools
```

### Q3: Top 10 các trường có điểm tổng hợp SAT cao nhất
### Viết đoạn code xuất ra danh sách 10 trường có điểm SAT tổng hợp cao nhất, lưu trong pandas DataFrame có tên top_10_schools  ( chứa "school_name" và cột mới có tên "total_SAT"), sắp xếp giảm dần theo cột "total_SAT".

```python
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
top_10_schools = schools.sort_values(by = "total_SAT", ascending = False).head(10)
top_10_schools = top_10_schools[["school_name", "total_SAT"]]
top_10_schools
```

### Q4: Quận (borough) nào có độ lệch chuẩn của điểm SAT tổng hợp lớn nhất
### Lưu kết quả trong pandas DataFrame tên largest_std_dev. DataFrame này chứa 1 hàng gồm các cột "borough", " building_code" , "average_SAT", std_SAT" . Giá trị độ lệch chuẩn làm tròn đến 2 chữ số thập phân.

```python
schools.head()
```

```python
# Tính độ lệch chuẩn (std) của từng quận
std_by_borough = schools.groupby('borough')['total_SAT'].std()
print(std_by_borough)

# Tìm quận có độ lệch chuẩn lớn nhất
largest_std_borough = std_by_borough.idxmax()
print("\nQuận có độ lệch chuẩn cao nhất:", largest_std_borough)

# Tính giá trị độ lệch chuẩn lớn nhất
largest_std = round(std_by_borough[largest_std_borough], 2) # Lam tròn số
print("Độ lệch chuẩn cao nhất:", largest_std)

# Tính điểm SAT trung bình của quận có độ lệch chuẩn lớn nhất
average_sat_borough = schools[schools['borough'] == largest_std_borough]['total_SAT'].mean()
print("Điểm SAT trung bình của quận:", average_sat_borough)

# Tìm mã tòa nhà của quận có độ lệch chuẩn lớn nhất
building_code = schools[schools['borough'] == largest_std_borough]['building_code'].iloc[0]
print("Mã tòa nhà của quận:", building_code)

# Tạo DataFrame
largest_std_dev = pd.DataFrame({
'borough': largest_std_borough,
'building_code': building_code,
'average_SAT': average_sat_borough,
'std_SAT': largest_std
}, index=[0])
print("\nDataFrame:")
largest_std_dev
```
