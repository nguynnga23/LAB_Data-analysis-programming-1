### BÀI 1 : KMEAN

LÝ THUYẾT

K-means (viết tắt của K-means clustering) là một thuật toán phân cụm dữ liệu không giám sát phổ biến được sử dụng để phân chia một tập dữ liệu thành một số cụm (k) dựa trên sự giống nhau của các điểm dữ liệu. 

1. Thuật toán này hoạt động bằng cách lặp lại các bước sau:
- Khởi tạo:
Chọn ngẫu nhiên k điểm dữ liệu làm trung tâm cụm ban đầu.
- Gán nhãn:
Gán mỗi điểm dữ liệu cho cụm có trung tâm gần nhất với điểm dữ liệu đó (theo khoảng cách Euclid).
- Cập nhật trung tâm:
Tính trung bình cộng của các điểm dữ liệu trong mỗi cụm để cập nhật vị trí của trung tâm cụm.
- Lặp lại:
Lặp lại bước 2 và 3 cho đến khi các nhãn cluster hoặc vị trí trung tâm cụm không thay đổi đáng kể.

2. Thuật toán K-means có một số ưu điểm như:

- Đơn giản: Dễ dàng hiểu và triển khai.
- Hiệu quả: Có thể xử lý tập dữ liệu lớn một cách hiệu quả.
- Có thể mở rộng: Có thể áp dụng cho nhiều loại dữ liệu khác nhau.
- Không có giả định: Không phụ thuộc vào các giả định cụ thể về phân bố dữ liệu hoặc cấu trúc cụm.

3. Tuy nhiên, K-means cũng có một số hạn chế:

- Kết quả phụ thuộc vào giá trị k: Việc lựa chọn giá trị k phù hợp là rất quan trọng và có thể ảnh hưởng đến kết quả phân cụm.
- Có thể nhạy cảm với nhiễu: Các điểm dữ liệu nhiễu có thể ảnh hưởng đến vị trí trung tâm cụm và kết quả phân cụm.
- Có thể gặp khó khăn với dữ liệu phi tuyến tính: Có thể gặp khó khăn khi phân cụm dữ liệu có cấu trúc phức tạp hoặc phi tuyến tính.


4. Ứng dụng của K-means:

- Phân tích thị trường: Phân chia khách hàng thành các nhóm dựa trên thói quen mua sắm hoặc sở thích.
- Phân tích sinh học: Phân loại gen hoặc protein dựa trên cấu trúc hoặc chức năng.
- Phát hiện gian lận: Phát hiện các giao dịch thẻ tín dụng gian lận.
- Xử lý ảnh: Phân chia hình ảnh thành các vùng đối tượng khác nhau.

5. Ví dụ về K-means:

- Hãy tưởng tượng bạn có một tập dữ liệu gồm các điểm đại diện cho vị trí của khách hàng tại một cửa hàng. Bạn muốn sử dụng K-means để phân chia khách hàng thành các nhóm dựa trên vị trí của họ trong cửa hàng.

- Khởi tạo: Chọn ngẫu nhiên 3 điểm dữ liệu làm trung tâm cụm ban đầu.
- Gán nhãn: Gán mỗi điểm dữ liệu cho cụm có trung tâm gần nhất với điểm dữ liệu đó.
- Cập nhật trung tâm: Tính trung bình cộng của các điểm dữ liệu trong mỗi cụm để cập nhật vị trí của trung tâm cụm.
- Lặp lại: Lặp lại bước 2 và 3 cho đến khi các nhãn cluster hoặc vị trí trung tâm cụm không thay đổi đáng kể.
- Kết quả của thuật toán K-means có thể cho bạn biết rằng có 3 nhóm khách hàng chính trong cửa hàng: nhóm khách hàng thường xuyên mua sắm ở khu vực phía trước cửa hàng, nhóm khách hàng thường xuyên mua sắm ở khu vực phía sau cửa hàng và nhóm khách hàng thường xuyên mua sắm ở khu vực trung tâm cửa hàng.

VẬN DỤNG

Đầu vào : Dữ liệu X và số lượng cluster cần tìm K.

Đầu ra : Các center M và label vector cho từng điểm dữ liệu Y.

1. Chọn K điểm bất kỳ làm các center ban đầu.
2. Phân mỗi điểm dữ liệu vào cluster có center gần nó nhất.
3. Nếu việc gán dữ liệu vào từng cluster ở bước 2 không thay đổi so với vòng lặp trước nó thì ta dừng
thuật toán.
4. Cập nhật center cho từng cluster bằng cách lấy trung bình cộng của tất các các điểm dữ liệu đã
được gán vào cluster đó sau bước 2.
5. Quay lại bước 2.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans. fit(random_array)

# Trích xuất các trung tâm của các cụm và nhãn của mỗi mẫu
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Vẽ biểu đồ Scatter plot của dữ liệu và các trung tâm cụm
plt.figure(figsize=(8, 6))
plt. scatter(random_array[:, 0], random_array[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='red', s=300)
plt.title('Biểu đồ Scatter plot của dữ liệu và các trung tâm cụm')
plt.xlabel('Trục X')
plt.ylabel('Trục Y')
plt.grid(True)
plt.show()
```

### BÀI 2 : CHỌN k = ? LÀ PHÙ HỢP

```python
from sklearn.cluster import KMeans
k_values = range(1, 11)

# Calculate inertia for each K value
inertia_values = []
for k in k_values:
    kmeans = KMeans (n_clusters=k)
    kmeans.fit(random_array)
    inertia_values.append(kmeans.inertia_)

# Plot inertia values for different K values
plt.plot(k_values, inertia_values, marker='o')
plt.title('Inertia vs. Number of Clusters')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks (k_values)
plt.grid(True)
plt.show()
```

KẾT LUẬN : Thuật toán K-means Clustering được áp dụng cho dữ liệu có hai đặc trưng. Đồ thị hiển thị quán tính (độ biến đổi trong cụm) cho các giá trị k khác nhau. Điểm "khuỷu tay", nơi đường cong bắt đầu phẳng lại ở khoảng **k = 4**, cho thấy k = 4 có thể là lựa chọn phù hợp cho tập dữ liệu này. Tuy nhiên, có thể cần phân tích và đánh giá thêm với các giá trị k khác nhau để xác nhận số lượng cụm tối ưu.

LÝ THUYẾT :
Lý do chọn điểm khuỷu tay trong phương pháp Elbow Method (Phương pháp Khuỷu tay) cho thuật toán K-means Clustering:

1. Hiểu biết trực quan:
- Phương pháp Elbow Method cung cấp một cách trực quan để lựa chọn giá trị k tối ưu cho thuật toán K-means Clustering.
- Khi vẽ đồ thị độ biến đổi trong cụm (WCSS) hoặc silhouette score theo các giá trị k khác nhau, ta có thể quan sát sự thay đổi của các chỉ số này.
- Điểm khuỷu tay (elbow point) là điểm mà đường cong bắt đầu phẳng hoặc giảm mạnh.

2. Giảm thiểu độ biến đổi trong cụm:

- Giá trị k trước điểm khuỷu tay tương ứng với việc tăng số lượng cụm không làm giảm đáng kể độ biến đổi trong cụm.
- Nói cách khác, việc thêm quá nhiều cụm không mang lại lợi ích đáng kể trong việc phân biệt các nhóm dữ liệu.
- Sau điểm khuỷu tay, độ biến đổi trong cụm giảm chậm hơn, cho thấy việc thêm cụm mới không cải thiện đáng kể chất lượng phân cụm.


3. Tối ưu hóa hiệu quả tính toán:

- Lựa chọn k tại điểm khuỷu tay giúp cân bằng giữa hiệu quả tính toán và chất lượng phân cụm.
- Việc sử dụng quá nhiều cụm có thể dẫn đến thời gian tính toán lâu hơn và kết quả phức tạp hơn cần giải thích.
- Ngược lại, sử dụng quá ít cụm có thể bỏ qua các nhóm dữ liệu riêng biệt, dẫn đến kết quả không chính xác.


4. Phù hợp với nhiều loại dữ liệu:

- Phương pháp Elbow Method tương đối đơn giản và dễ áp dụng cho nhiều loại dữ liệu khác nhau.
- Nó không phụ thuộc vào các giả định cụ thể về phân bố dữ liệu hoặc cấu trúc cụm.


5. Hỗ trợ lựa chọn ban đầu:

- Mặc dù điểm khuỷu tay không phải lúc nào cũng cung cấp giá trị k chính xác nhất, nó có thể là một điểm khởi đầu hữu ích để lựa chọn k ban đầu.
- Sau đó, có thể sử dụng các phương pháp đánh giá khác (ví dụ: silhouette score) để tinh chỉnh lựa chọn k.
