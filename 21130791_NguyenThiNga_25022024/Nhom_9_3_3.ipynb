```python
top_counts = coffee_ratings['Country of Origin'].value_counts()
top_counts.head(5)
```

```python
coffee_ratings.info()
```

```python
top_counted_countries  = ["Taiwan", "Guatemala", "Colombia", "Honduras", "Thailand"]
top_counted_subset = coffee_ratings['Country of Origin'].isin(top_counted_countries)
coffe_ratings_top = coffee_ratings[top_counted_subset]
```

```python
coffee_ratings_samp = coffe_ratings_top.sample(frac=0.1, random_state=2021)
coffee_ratings_samp['Country of Origin'].value_counts(normalize = True)
```

```python
coffee_ratings['Country of Origin'].value_counts(normalize = True).head(4)
```

```python
# -> Không đại diện cho popul về mặt tỉ lệ
```

```python
# Proportional startified sampling
coffee_ratings_strat = coffe_ratings_top.groupby('Country of Origin')\
                      .sample(frac = 0.1, random_state = 2020)
```

```python
coffee_ratings_strat['Country of Origin'].value_counts(normalize = True)
```

```python
# -> tăng về tỷ lệ %, kì vọng sẽ đại diện cho popul
```

```python
# Equal counts stratified sampling
coffee_ratings_eq = coffe_ratings_top.groupby('Country of Origin')\
                    .sample(n=10, random_state=2020)
```

```python
coffee_ratings_eq['Country of Origin'].value_counts(normalize=True)
```

# Tỷ lệ trong sample = nhau

```python
# Weighted random sampling
coffee_ratings_weight = coffe_ratings_top
condition = coffee_ratings_weight['Country of Origin'] == "Taiwan"
coffee_ratings_weight['weight'] = np.where(condition, 2, 1)
coffee_ratings_weight = coffee_ratings_weight.sample(frac = 0.1, weights = 'weight')
```

```python
coffee_ratings_weight['Country of Origin'].value_counts(normalize=True)
```

Nhận xét: cho ra sample có tỷ lệ gần hơn với tỷ lệ của population

```python
# Srattified sampling
varieties_pop = list(coffee_ratings['Variety'].unique())
```

```python
# Step 1
import random
varieties_samp = random.sample(varieties_pop, k = 3)
```

```python
# Step 2
variety_cond = coffee_ratings['Variety'].isin(varieties_samp)
coffee_ratings_cluster = coffee_ratings[variety_cond]
coffee_ratings_cluster['Variety'] = coffee_ratings_cluster['Variety'].astype('category')
```

```python
coffee_ratings_cluster['Variety'] = coffee_ratings_cluster['Variety'].cat.remove_unused_categories()
```

```python
coffee_ratings_cluster['Variety'] = coffee_ratings_cluster['Variety'].astype('object')
coffee_ratings_cluster.groupby('Variety')\
      .sample(n=1, random_state = 1)
```
